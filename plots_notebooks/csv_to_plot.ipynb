#  diff/same/hps/lps test on model trained on frontal faces

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/home/administrator/experiments/high_importance_features_full_dataset/vgg16/results/comparisons.csv')
title = '1000_ids_300_img_per_id_2'
# df = pd.read_csv(r"/home/administrator/experiments/500_ids_300_img_per_id_2/vgg16/results/100_ids_non_frontal_300_per_id_comparison.csv")
pd.set_option('display.max_rows', df.shape[0]+1)

print(df)
```

```python
# change unnamed column to 'pairs'
df = df.rename(columns = {'Unnamed: 0':'pairs'})
df
```

```python
#split the tuples to two columns
df_splited = df['pairs'].str.split(',', expand=True)

#adding   
df.insert(loc=0, column='img1', value=df_splited[0] )
df.insert(loc=1, column='img2', value=df_splited[1] )

  
# Dropping old Name columns 
df.drop(columns =["pairs"], inplace = True) 
```

```python
df
```

```python
#reorder columns by layer order
df = df[['img1', 'img2', 'input', 'conv1', 'conv2', 'conv3', 'conv4', 'conv5', 'fc6', 'fc7', 'fc8','type' ]]
```

```python
df
```

```python
#remove all last rows - save only DIFF,HPS,LPS,SAME type.

df = df[~df.type.str.contains("frontal")]
```

```python
df
```

```python
max_value_df = df.groupby(['type']).max()
max_value_df
```

```python
df = df.reset_index(drop=True)
df
```

```python
max_value_df = max_value_df.loc[np.repeat(max_value_df.index.values, 25)]
max_value_df = max_value_df.reset_index(drop=True)
max_value_df
```

```python
#remove imgs name
df = df.drop(df.columns[[0,1]], axis=1)
#make the max value table
max_value_df = max_value_df.drop(max_value_df.columns[[0,1]], axis=1)
max_value_df
```

```python
df
```

```python
#remove the "type" column in order to make a total float table
df = df.drop('type', axis=1)
```

```python
df
```

```python
divided_df = df.div(max_value_df)
divided_df
```

```python
list_type = ['diff']*25 + ['HPS']*25+ ['LPS']*25 + ['same']*25
```

```python
divided_df.insert(0, 'list_type', list_type, allow_duplicates = False)
```

```python
divided_df
```

```python
#mean of each columns by type of pairs
means_df = divided_df.groupby(['list_type']).mean()
means_df = means_df.rename_axis("layers", axis="columns")
means_df
```

```python
means_df = means_df.transpose()
means_df
```

```python
means_df = means_df.reindex(['input', 'conv1', 'conv2', 'conv3', 'conv4', 'conv5', 'fc6', 'fc7', 'fc8' ])
means_df
```

```python
plot_kwargs={'grid':True,'figsize':(20,8),
 'fontsize':12}
```

```python
plot = means_df.plot(**plot_kwargs, title = title, ylim = [0,1])
```
