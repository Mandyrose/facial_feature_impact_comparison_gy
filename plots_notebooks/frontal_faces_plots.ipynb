```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/home/administrator/experiments/500_ids_300_img_per_id_2/vgg16/results/comparisons.csv')
pd.set_option('display.max_rows', df.shape[0]+1)

print(df)
```

```python
# change unnamed column to 'pairs'
df = df.rename(columns = {'Unnamed: 0':'pairs'})
df
```

```python
#split the tuples to two columns
df_splited = df['pairs'].str.split(',', expand=True)

#adding   
df.insert(loc=0, column='img1', value=df_splited[0] )
df.insert(loc=1, column='img2', value=df_splited[1] )

  
# Dropping old Name columns 
df.drop(columns =["pairs"], inplace = True) 
```

```python
df
```

```python
#remove all last rows - save only DIFF,HPS,LPS,SAME type.

df = df[~df.type.str.contains("frontal")]
```

```python
df
```

```python
max_value_df = df.groupby(['type']).max()
max_value_df
```

```python
df = df.reset_index(drop=True)
df
```

```python
max_value_df = max_value_df.loc[np.repeat(max_value_df.index.values, 25)]
max_value_df = max_value_df.reset_index(drop=True)
max_value_df
```

```python
df = df.drop(df.columns[[0,1]], axis=1)
max_value_df = max_value_df.drop(max_value_df.columns[[0,1]], axis=1)
max_value_df
```

```python
df
```

```python
df = df.drop(df.columns[[-1]], axis=1)
```

```python
divided_df = df.div(max_value_df)
divided_df
```

```python
list_type = ['diff']*25 + ['HPS']*25+ ['LPS']*25 + ['same']*25
```
